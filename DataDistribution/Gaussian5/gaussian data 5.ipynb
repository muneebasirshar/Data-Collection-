```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D
import math as mt


#distribution parameters
mu1 = np.array([3, 2]).reshape(-1,1)
Sigma1 = np.array([[10 , 2], [2,  9]]).T
mu2=np.array([8,14]).reshape(-1,1)
Sigma2=np.array([[8,2],[2,10]]).T
mu3=np.array([13,14]).reshape(-1,1)
Sigma3=np.array([[9,1],[1,8]]).T
mu4=np.array([4,2]).reshape(-1,1)
Sigma4=np.array([[7,3],[3,8]]).T
mu5=np.array([10,15]).reshape(-1,1)
Sigma5=np.array([[11,3],[3,10]]).T

#location of Sensor
loc=np.array([15,15]).reshape(-1,1)

#calculation of Gaussian Function for given sensor location
# First Gaussian
dif1=loc-mu1
dif_transpose1=dif1.reshape(1,-1)
Sigma_det1 = np.linalg.det(Sigma1)
Sigma_inv1 = np.linalg.inv(Sigma1)
D1 = np.sqrt((2*np.pi)**2 * Sigma_det1)
fac1_1=np.matmul(dif_transpose1,Sigma_inv1)
fac2_1=np.matmul(fac1_1,dif1)
value1=np.exp(-fac2_1/2)/D1
# Second Gaussian
dif2=loc-mu2
dif_transpose2=dif2.reshape(1,-1)
Sigma_det2 = np.linalg.det(Sigma2)
Sigma_inv2 = np.linalg.inv(Sigma2)
D2= np.sqrt((2*np.pi)**2 * Sigma_det2)
fac1_2=np.matmul(dif_transpose2,Sigma_inv2)
fac2_2=np.matmul(fac1_2,dif2)
value2=np.exp(-fac2_2/2)/D2
# Third Gaussian
dif3=loc-mu3
dif_transpose3=dif3.reshape(1,-1)
Sigma_det3 = np.linalg.det(Sigma3)
Sigma_inv3 = np.linalg.inv(Sigma3)
D3 = np.sqrt((2*np.pi)**2 * Sigma_det3)
fac1_3=np.matmul(dif_transpose3,Sigma_inv3)
fac2_3=np.matmul(fac1_3,dif3)
value3=np.exp(-fac2_3/2)/D3
# Fourth Gaussian
dif4=loc-mu4
dif_transpose4=dif4.reshape(1,-1)
Sigma_det4 = np.linalg.det(Sigma4)
Sigma_inv4 = np.linalg.inv(Sigma4)
D4 = np.sqrt((2*np.pi)**2 * Sigma_det4)
fac1_4=np.matmul(dif_transpose4,Sigma_inv4)
fac2_4=np.matmul(fac1_4,dif4)
value4=np.exp(-fac2_4/2)/D4
# Fifth Gaussian
dif5=loc-mu5
dif_transpose5=dif5.reshape(1,-1)
Sigma_det5 = np.linalg.det(Sigma5)
Sigma_inv5 = np.linalg.inv(Sigma5)
D5 = np.sqrt((2*np.pi)**2 * Sigma_det5)
fac1_5=np.matmul(dif_transpose5,Sigma_inv5)
fac2_5=np.matmul(fac1_5,dif5)
value5=np.exp(-fac2_5/2)/D5


#initializing Distributionns
dist1=[]
dist2=[]
dist3=[]
dist4=[]
dist5=[]

#Adding Spatial and Temporal characteristics

# Dist 1
for i in range(0, 200):
    if (i%2 == 0):
        j1=i/10
        d1=mt.e**(-j1/10)
        dampingfac1=value1*d1
        cosinefac1=mt.cos(2*mt.pi*0.1*j1)
        a=value1*dampingfac1*cosinefac1
        dist1.append(a[0,0])
         
 
 
#Dist 2
for i in range(0, 200):
    if (i%2 == 0):
        j2=i/10
        d2=mt.e**(-j2/15)
        dampingfac2=value2*d2
        cosinefac2=mt.cos(2*mt.pi*0.2*j2)
        b=value2*dampingfac2*cosinefac2
        dist2.append(b[0,0])
         
            
# Dist 3
for i in range(0, 200):
    if (i%2 == 0):
        j3=i/10
        d3=mt.e**(-j3/8)
        dampingfac3=value3*d3
        cosinefac3=mt.cos(2*mt.pi*0.05*j3)
        c=value3*dampingfac3*cosinefac3
        dist3.append(c[0,0])
         
            
            
# Dist 4
for i in range(0, 200):
    if (i%2 == 0):
        j4=i/10
        d4=mt.e**(-j4/8)
        dampingfac4=value4*d4
        cosinefac4=mt.cos(2*mt.pi*0.3*j4)
        c=value4*dampingfac4*cosinefac4
        dist4.append(c[0,0])
         
            
# Dist 5
for i in range(0, 200):
    if (i%2 == 0):
        j5=i/10
        d5=mt.e**(-j5/8)
        dampingfac5=value5*d5
        cosinefac5=mt.cos(2*mt.pi*0.4*j5)
        c=value5*dampingfac5*cosinefac5
        dist5.append(c[0,0])
            
#summing up the distributions
SummingTwo=np.add(dist1,dist2)
Summing3=np.add(SummingTwo,dist3)
Summing4=np.add(Summing3,dist4)
SummingAll=np.add(Summing4,dist5)
SensorValu=np.multiply(1000,SummingAll)
len(SensorValu)
max(SensorValu)
np.savetxt("ID64-gauss5.csv", SensorValu, delimiter=",")
 
```
