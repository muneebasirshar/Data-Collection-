```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
DataFile = pd.read_csv('C:/Users/muneeba/Desktop/THESIS/DataDistribution/Gaussian3/gauss3/DataDistributionPCA.csv')
X=DataFile.iloc[:,0] # sensor IDs
X=X[:64]
Y1=pd.read_csv('C:/Users/muneeba/Desktop/THESIS/DataDistribution/Gaussian5/Gauss5.csv')
Y1=Y1[0:64]
Y1=Y1.iloc[:,0:64]
```

```python
Y=np.mean(Y1,axis=0) #since we have a large time span we find mean for each sensor
np.shape(Y)
```

```python
# Visualize our dataset
plt.scatter(X, Y,  color='blue')
plt.xlabel('ID of the sensors')
plt.ylabel('Mean Values')
plt.title('Data Distribution of Sensors Data for 5 Gaussian')
plt.savefig('Data Mean Distribution')
plt.show()
```
